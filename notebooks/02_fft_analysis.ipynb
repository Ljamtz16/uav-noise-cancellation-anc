# 02 - FFT Analysis
Calcula FFT y detecta frecuencia dominante de la señal UAV.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))
```

```python
from src.signal.generate_signal import generate_uav_signal
from src.analysis.fft_analysis import compute_fft, dominant_frequency
from src.utils.plotting import plot_spectrum

fs = 8000
_, x, f_bpf = generate_uav_signal(fs=fs, duration=2, rpm=6000, n_blades=2)
freq, mag = compute_fft(x, fs)
f_dom = dominant_frequency(x, fs, min_hz=20)

print(f'BPF teórica: {f_bpf:.2f} Hz')
print(f'Frecuencia dominante detectada: {f_dom:.2f} Hz')
plot_spectrum(freq, mag, title='FFT - UAV signal')
```
